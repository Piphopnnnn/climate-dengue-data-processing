```python
import pandas as pd

# 1. โหลดข้อมูลทั้งสองไฟล์
dengue_df = pd.read_csv(r"D:\Project_Dengue\DHF\DHF_clean\Surin.csv")
weather_df = pd.read_csv(r"D:\Project_Dengue\rain_data\Surin\สถานีอุตุนิยมวิทยาสุรินทร์\rain_monthly_Surin_สถานีอุตุนิยมวิทยาสุรินทร์.csv")

# 2. แปลงคอลัมน์วันที่ให้เป็นรูปแบบ datetime
dengue_df['Date'] = pd.to_datetime(dengue_df['Date'])
weather_df['time'] = pd.to_datetime(weather_df['time'])

# 3. สร้าง Key สำหรับรวมข้อมูลโดยใช้ Year-Month (เพื่อให้ข้อมูลเดือนเดียวกันมาเจอกัน)
dengue_df['month_key'] = dengue_df['Date'].dt.to_period('M')
weather_df['month_key'] = weather_df['time'].dt.to_period('M')

# 4. รวมข้อมูล (Merge) เข้าด้วยกัน
merged_df = pd.merge(dengue_df, weather_df, on='month_key', how='inner')

# 5. ลบคอลัมน์ที่ไม่ได้ใช้และจัดระเบียบใหม่
merged_df = merged_df.drop(columns=['month_key', 'time'])

# 6. บันทึกไฟล์ใหม่
merged_df.to_csv(r"D:\Project_Dengue\FINAL_DATA\DHF\เขตสุขภาพ 9\Surin_สถานีอุตุนิยมวิทยาสุรินทร์_DHF_weather_merged_monthly.csv", index=False)

print("รวมข้อมูลสำเร็จ! ไฟล์ถูกบันทึกในชื่อ 'merged_dengue_weather_data.csv'")
```
